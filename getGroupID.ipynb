```python
# Debug script untuk Microsoft Graph API Authentication
import msal
import requests
from internal_assistant_core import settings

def debug_auth_configuration():
    """Debug konfigurasi autentikasi"""
    print("=== DEBUG MICROSOFT GRAPH AUTH ===")
    
    # 1. Cek konfigurasi settings
    print("\n1. Settings Configuration:")
    print(f"MS_CLIENT_ID: {'✓ Set' if settings.MS_CLIENT_ID else '✗ Missing'}")
    print(f"MS_CLIENT_SECRET: {'✓ Set' if settings.MS_CLIENT_SECRET else '✗ Missing'}")
    print(f"MS_TENANT_ID: {'✓ Set' if settings.MS_TENANT_ID else '✗ Missing'}")
    print(f"MS_GROUP_ID: {'✓ Set' if settings.MS_GROUP_ID else '✗ Missing'}")
    print(f"MS_GRAPH_SCOPE: {settings.MS_GRAPH_SCOPE}")
    print(f"Authority: {settings.ms_authority}")
    
    # 2. Test MSAL App Creation
    print("\n2. MSAL App Creation:")
    try:
        app = msal.ConfidentialClientApplication(
            settings.MS_CLIENT_ID,
            authority=settings.ms_authority,
            client_credential=settings.MS_CLIENT_SECRET,
        )
        print("✓ MSAL App created successfully")
    except Exception as e:
        print(f"✗ MSAL App creation failed: {e}")
        return
    
    # 3. Test Token Acquisition
    print("\n3. Token Acquisition:")
    try:
        result = app.acquire_token_for_client(scopes=[settings.MS_GRAPH_SCOPE])
        
        if "access_token" in result:
            print("✓ Token acquired successfully")
            print(f"Token type: {result.get('token_type', 'N/A')}")
            print(f"Expires in: {result.get('expires_in', 'N/A')} seconds")
            # Don't print actual token for security
            print(f"Token length: {len(result['access_token'])} characters")
            
            # 4. Test basic Graph API call
            print("\n4. Test Graph API Call:")
            token = result["access_token"]
            test_graph_api(token)
            
        else:
            print("✗ Token acquisition failed")
            print(f"Error: {result.get('error', 'Unknown')}")
            print(f"Error description: {result.get('error_description', 'N/A')}")
            print(f"Correlation ID: {result.get('correlation_id', 'N/A')}")
            
            # Common issues troubleshooting
            suggest_fixes(result)
            
    except Exception as e:
        print(f"✗ Exception during token acquisition: {e}")

def test_graph_api(token: str):
    """Test Graph API dengan token yang diperoleh"""
    headers = {"Authorization": f"Bearer {token}"}
    
    # Test 1: Basic Graph API call (me endpoint won't work for app-only, use /groups)
    print("Testing Graph API calls...")
    
    # Test groups endpoint
    try:
        url = f"https://graph.microsoft.com/v1.0/groups/{settings.MS_GROUP_ID}"
        response = requests.get(url, headers=headers)
        
        if response.status_code == 200:
            print("✓ Groups API call successful")
            group_data = response.json()
            print(f"Group name: {group_data.get('displayName', 'N/A')}")
        else:
            print(f"✗ Groups API call failed: {response.status_code}")
            print(f"Response: {response.text}")
            
    except Exception as e:
        print(f"✗ Groups API exception: {e}")
    
    # Test planner plans endpoint
    try:
        url = f"https://graph.microsoft.com/v1.0/groups/{settings.MS_GROUP_ID}/planner/plans"
        response = requests.get(url, headers=headers)
        
        if response.status_code == 200:
            print("✓ Planner API call successful")
            plans = response.json().get('value', [])
            print(f"Found {len(plans)} plans")
            for plan in plans[:3]:  # Show first 3
                print(f"  - {plan.get('title', 'Untitled')}")
        else:
            print(f"✗ Planner API call failed: {response.status_code}")
            print(f"Response: {response.text}")
            analyze_api_error(response)
            
    except Exception as e:
        print(f"✗ Planner API exception: {e}")

def analyze_api_error(response):
    """Analisis detail error dari Graph API"""
    try:
        error_data = response.json()
        error_code = error_data.get('error', {}).get('code', 'Unknown')
        error_message = error_data.get('error', {}).get('message', 'Unknown')
        
        print(f"\nDetailed Error Analysis:")
        print(f"Error Code: {error_code}")
        print(f"Error Message: {error_message}")
        
        # Specific error handling
        if error_code == "Forbidden":
            print("\n🔍 PERMISSION ISSUE DETECTED:")
            print("- App registration mungkin tidak memiliki permission yang cukup")
            print("- Perlu permission: Group.Read.All, Tasks.Read, Tasks.ReadWrite")
            print("- Admin consent mungkin belum diberikan")
            
        elif error_code == "InvalidAuthenticationToken":
            print("\n🔍 TOKEN ISSUE DETECTED:")
            print("- Token mungkin expired atau invalid")
            print("- Cek CLIENT_ID dan CLIENT_SECRET")
            print("- Pastikan app registration aktif")
            
        elif error_code == "NotFound":
            print("\n🔍 RESOURCE NOT FOUND:")
            print("- Group ID mungkin salah atau tidak ada")
            print("- User/App tidak memiliki akses ke group tersebut")
            
    except:
        print("Could not parse error response")

def suggest_fixes(auth_result):
    """Saran perbaikan berdasarkan error autentikasi"""
    error = auth_result.get('error', '')
    
    print(f"\n🔧 SUGGESTED FIXES:")
    
    if error == "invalid_client":
        print("- Cek CLIENT_ID dan CLIENT_SECRET di Azure App Registration")
        print("- Pastikan tidak ada typo atau extra spaces")
        print("- Verify app registration masih aktif")
        
    elif error == "invalid_scope":
        print("- Cek scope permission di app registration")
        print("- Pastikan ada permission untuk: https://graph.microsoft.com/.default")
        print("- Grant admin consent untuk permissions")
        
    elif error == "unauthorized_client":
        print("- App registration mungkin tidak support client credentials flow")
        print("- Cek authentication flow settings di Azure")
        
    else:
        print("- Cek semua konfigurasi di Azure App Registration")
        print("- Verify tenant ID benar")
        print("- Pastikan app registration memiliki required permissions")

def check_required_permissions():
    """List permission yang dibutuhkan"""
    print("\n📋 REQUIRED PERMISSIONS:")
    print("Di Azure App Registration, pastikan ada permission berikut:")
    print("Application Permissions (tidak perlu user login):")
    print("- Group.Read.All")
    print("- Tasks.Read") 
    print("- Tasks.ReadWrite")
    print("- Directory.Read.All (optional)")
    print("\nDAN pastikan Admin Consent sudah diberikan!")

if __name__ == "__main__":
    debug_auth_configuration()
    check_required_permissions()
    
    print("\n" + "="*50)
    print("NEXT STEPS:")
    print("1. Jalankan script ini untuk melihat error detail")
    print("2. Perbaiki konfigurasi berdasarkan hasil debug")
    print("3. Test kembali API call")
    print("="*50)
```
